```python
import networkx as nx
import matplotlib.pyplot as plt

# Create simple Research Supervisor list
supervisor = ['Dr. Ansah', 'Prof. Solo', 'Dr. PBS', 'Prof. Aseidu']
student= ['George', 'David','Prince','Princilla']

# Create simple colaboration
collaborations = [
   ('Dr. Ansah', 'George'),
    ('Prof. Solo', 'David'),
    ('Dr. PBS', 'Prince'),
    ('Dr. Ansah', 'Princilla'),
    ('Prof. Solo', 'George'),
    ('Dr. PBS', 'David'),
    ('Dr. Ansah', 'Prince'),
]

```

```python
#create graph
G = nx.DiGraph()

# Add nodes
G.add_nodes_from(supervisor)
G.add_nodes_from(student)

# Tadd edges
G.add_edges_from(collaborations)
```

```python
# Basic info
print(f"Total people: {G.number_of_nodes()}")
print(f"Total relationships: {G.number_of_edges()}")
```

```python
# Who has how many connections
for person in G.nodes():
    connections = G.degree(person)
    print(f"{person}: {connections}")
```

```python
# Check isolated people
isolated = [person for person in G.nodes() if G.degree(person) == 0]
if isolated:
    print(isolated)
else:
    print("None")
```

```python
#made isolated nodes red
node_colors = ['red' if node in isolated else 'lightblue' for node in G.nodes()]
```

```python
# Ploting the graph
plt.figure(figsize=(10, 8))
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=3000)
nx.draw_networkx_edges(G, pos, arrows=True, arrowsize=20, arrowstyle='->')
nx.draw_networkx_labels(G, pos, font_size=10)
plt.axis('off')
plt.show
```
